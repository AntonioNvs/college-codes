## Modelo

```python
valid_index = X_dummies.reset_index().merge(features_voronoi, left_on="index", right_on="instance_index")["index"]
```

```python
X_dummies = X_dummies.loc[valid_index]
X_features = X_dummies.drop(["pass_outcome_type", "pass_high_point_type"], axis=1).values
y = (X_dummies["pass_outcome_type"] == "D").astype(np.int32).values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, brier_score_loss

X_train, X_test, y_train, y_test = train_test_split(
    X_features, y, test_size=0.2, random_state=42
)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100, random_state=42)
forest.fit(X_train, y_train)
```

```python
y_prob_forest = forest.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, forest.predict(X_test))
forest_roc_auc = roc_auc_score(y_test, y_prob_forest)
brier_score = brier_score_loss(y_test, y_prob_forest)

print(f'Accuracy: {accuracy}, ROC AUC: {forest_roc_auc}')
print(f'Brier Score: {brier_score}')
```

### XGBoost

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
```

### MLP

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def generate_mlp_model(input_shape):
  model = keras.Sequential(
      [
          layers.Dense(64, activation="selu", input_shape=input_shape),
          layers.Dropout(0.2),
          layers.Dense(64, activation="selu"),
          layers.Dropout(0.2),
          layers.Dense(32, activation="tanh"),
          layers.Dense(1, activation="sigmoid")
      ]
  )

  model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

  return model
```

```python
model = generate_mlp_model((X_train.shape[1], ))

model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=128
)
```

```python
# Calculando acurÃ¡cia e ROC

y_prob_mlp = model.predict(X_test)

accuracy = accuracy_score(y_test, y_prob_mlp > 0.5)
mlp_roc_auc = roc_auc_score(y_test, y_prob_mlp)
brier_score = brier_score_loss(y_test, y_prob_mlp)

print(f'Accuracy: {accuracy}, ROC AUC: {mlp_roc_auc}')
print(f'Brier Score: {brier_score}')
```

```python
selected_voronoi = X_dummies.reset_index().merge(features_voronoi, left_on="index", right_on="instance_index")

X_voronoi = selected_voronoi[features_voronoi.columns].drop(["instance_index", "alpha_angle"], axis=1)
X_voronoi["label"] = (selected_voronoi["pass_outcome_type"] == "D").astype(np.int32).values
```

```python
X_voronoi_features = pd.get_dummies(X_voronoi, columns=["adversary_domain_pass_count"])

# Removendo instÃ¢ncias nulas
X_voronoi_features.dropna(inplace=True)

# Transformando bool em int
for col in X_voronoi_features.columns:
    if X_voronoi_features[col].dtype == bool:
        X_voronoi_features[col] = X_voronoi_features[col].astype(int)
```

```python
from sklearn.utils import resample

X_minority = X_voronoi_features.loc[X_voronoi_features["label"] == 1, :]

X_minority_upsampled, _ = resample(
    X_minority,
    X_voronoi_features[X_voronoi_features["label"] == 1]["label"],
    replace=True,  
    n_samples=len(X_voronoi_features) - len(X_minority),
    random_state=123
)

X_voronoi_features_upsampled = pd.concat([X_voronoi_features.loc[X_voronoi_features["label"] == 0, :], X_minority_upsampled])
```

```python
X_final = X_voronoi_features_upsampled.drop("label", axis=1)
y_final = X_voronoi_features_upsampled["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_final, test_size=0.2, random_state=42
)
```

```python
# Normalizando
features_not_dummy = X_voronoi.drop(["adversary_domain_pass_count", "label"], axis=1).columns

mean = X_train[features_not_dummy].mean()
std = X_train[features_not_dummy].std()

X_test.loc[:, features_not_dummy] = (X_test[features_not_dummy] - mean) / std
X_train.loc[:, features_not_dummy] = (X_train[features_not_dummy] - mean) / std
```

```python
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42, max_features="sqrt"),
    "XGBoost": xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    # "MLP": generate_mlp_model((X_train.shape[1],))
}
```

```python
results = []

for name, model in models.items():
  print(f"Evaluando {name}..")

  if name == "MLP":
    model.fit(X_train, y_train, epochs=10, batch_size=128)
    y_prob = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_prob > 0.5)
    roc_auc = roc_auc_score(y_test, y_prob)
    brier_score = brier_score_loss(y_test, y_prob)

  else:
    model.fit(X_train, y_train)

    y_prob = model.predict_proba(X_test)[:, 1]

    accuracy = accuracy_score(y_test, model.predict(X_test))
    roc_auc = roc_auc_score(y_test, y_prob)
    brier_score = brier_score_loss(y_test, y_prob)

  results.append({
      "name": name,
      "accuracy": accuracy,
      "roc_auc": roc_auc,
      "brier_score": brier_score,
      "y_prob": y_prob,
      "model": model
  })
```
